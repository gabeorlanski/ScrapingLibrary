```python
import json
from collections import defaultdict
cat = json.loads(open("test_data.json").read())["categories"]["section"]
data = defaultdict(dict)
```

```python
for c in cat.keys():
    if c != "num_results" and c != "num_children":
        tmp = cat[c]["children"][0]
        parent_section = cat[c]["section"]
        print(parent_section)
        if parent_section not in data:
            data[parent_section] = {}
        for k in tmp.keys():
            try:
                del tmp[k]["children"]
            except:
                pass
            if "l2" in k:
                try:
                    l2_section,l1_section = tmp[k]["l2_subsection"].split(" - ")
                    data[parent_section][l1_section][l2_section] = tmp[k]
                except:
                    try:
                        l2_section,l1_section = tmp[k]["l2_subsection"].split(" - ")
                        data[parent_section][l1_section] = {l2_section:tmp[k]}
                    except:
                        pass
            
                
                
```

```python
with open("categories.json","w") as f:
    f.write(json.dumps(data, indent=4, sort_keys=True))
```
